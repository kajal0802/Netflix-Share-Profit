```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nflx = pd.read_csv('NFLX.csv' , index_col = 'Date')
nflx['MA10'] = nflx['Close'].rolling(10).mean()
nflx['MA50'] = nflx['Close'].rolling(50).mean()
nflx = nflx.dropna()
nflx.head()
```

```python
plt.figure(figsize=(12, 8))
nflx['MA10'].plot(label='MA10')
nflx['MA50'].plot(label='MA50')
nflx['Close'].plot(label='Close')
plt.legend()
plt.show()
```

```python
nflx['Shares'] = [1 if nflx.loc[x, 'MA10']> nflx.loc[x, 'MA50'] else 0 for x in nflx.index]
```

```python
nflx['Close1'] = nflx['Close'].shift(-1)
nflx['Profit'] = [nflx.loc[x, 'Close1'] - nflx.loc[x, 'Close'] if nflx.loc[x, 'Shares']==1 else 0 for x in nflx.index]
nflx['wealth'] = nflx['Profit'].cumsum()
nflx.tail()
```

```python
nflx['wealth'].plot()
plt.title('Total money gain is {}'.format(nflx.loc[nflx.index[-2],'wealth']))
```
